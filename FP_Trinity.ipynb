```python
import pandas as pd

# Load datasets
hospitals_df = pd.read_csv("Hospitals.csv")
death_df = pd.read_csv("Number_of_Death.csv")
specialty_df = pd.read_csv("Specialty.csv")
deprivation_df = pd.read_csv("Deprivation.csv")
```

```python
# rename for consistency
death_df = death_df.rename(columns={"HBT": "HB"})
```

```python
# rename for consistency
hospitals_df = hospitals_df.rename(columns={"HealthBoard": "HB"})
```

```python
# Merge Deprivation and Death via Year and HB
merged_df = pd.merge(deprivation_df, death_df, on=["Year", "HB"], how="inner")
#merge hospitals
merged_df = pd.merge(merged_df, hospitals_df, on="HB", how="left")
```

```python
agg_specialty = specialty_df.groupby(["Year", "HB", "Location"]).agg({
    "Attendances": "sum",
    "DNAAppointments": "sum",
    "DNARate": "mean"
}).reset_index()
```

```python
merged_df = pd.merge(merged_df, agg_specialty, on=["Year", "HB", "Location"], how="left")
```
